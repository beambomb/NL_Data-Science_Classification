```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.metrics import f1_score
```

```python
train_df = pd.read_csv('/content/train (2).csv')
test_df = pd.read_csv('/content/test (2).csv')
```

```python
train_df
```

MEMBERSIHKAN DATASET TRAIN

```python
train_df
train_df['Jenis Kelamin'] = train_df['Jenis Kelamin'].replace(['M', 'F'], [1, 0])
train_df = train_df.drop(['Tempat lahir'], axis=1)
```

```python
test_df
```

MEMBERSIHKAN DATASET TEST

```python
test_id_for_submission = test_df['id']
```

```python
test_df = test_df.drop(['Tempat lahir', 'id'], axis=1)
test_df['Jenis Kelamin'] = test_df['Jenis Kelamin'].replace(['M', 'F'], [1, 0])
```

Pelatihan Model

```python
y_train = train_df['is_cholesterol_high']
x_train = train_df.drop(['is_cholesterol_high'], axis=1)
```

```python
x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
```

```python
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
val_predict = model.predict(x_val)
val_predict = pd.DataFrame(val_predict)
```

```python
y_val
```

```python
val_predict
```

```python
f1 = f1_score(y_val, val_predict, average='binary')
f1
```

TEST MODEL

```python
test_predict = model.predict(test_df)
test_predict
```

```python
submission_df = pd.DataFrame({
    'id': test_id_for_submission,
    'is_cholesterol_high': test_predict
})

submission_df.to_csv('submission.csv', index=False)
print("File submission.csv berhasil dibuat!")
```
